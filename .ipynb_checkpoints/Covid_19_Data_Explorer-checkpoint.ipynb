## Content

This data was scraped from woldometers.info on 2022-05-14 by Joseph Assaker.

225 countries are represented in this data.

All of countries have records dating from 2020-2-15 until 2022-05-14 (820 days per country).
That's with the exception of China, which has records dating from 2020-1-22 until 2022-05-14 (844 days per country), and Palau which has records dating from 2021-8-25 until 2022-05-14 (263 days per country)..

## Summary Data Columns Description:

country: designates the Country in which the the row's data was observed.

continent: designates the Continent of the observed country.

total_confirmed: designates the total number of confirmed cases in the observed country.

total_deaths: designates the total number of confirmed deaths in the observed country.

total_recovered: designates the total number of confirmed recoveries in the observed country.

active_cases: designates the number of active cases in the observed country.

serious_or_critical: designates the estimated number of cases in serious or critical conditions in the observed country.

total_cases_per_1m_population: designates the number of total cases per 1 million population in the observed country.

total_deaths_per_1m_population: designates the number of total deaths per 1 million population in the observed country.

total_tests: designates the number of total tests done in the observed country.

total_tests_per_1m_population: designates the number of total test done per 1 million population in the observed country.

population: designates the population count in the observed country.

## Daily Data Columns Description:

date: designates the date of observation of the row's data in YYYY-MM-DD format.

country: designates the Country in which the the row's data was observed.

cumulative_total_cases: designates the cumulative number of confirmed cases as of the row's date, for the row's country.

daily_new_cases: designates the daily new number of confirmed cases on the row's date, for the row's country.

active_cases: designates the number of active cases (i.e., confirmed cases that still didn't recover nor die) on the row's date, for the row's country.

cumulative_total_deaths: designates the cumulative number of confirmed deaths as of the row's date, for the row's country.

daily_new_deaths: designates the daily new number of confirmed deaths on the row's date, for the row's country.

## Data Sorting, Cleaning and Info 

**Import all libraries needed**

```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

**Load the dataset**

```python
df = pd.read_csv("corona_summary.csv")
```

**Check first 5 rows**

```python
df.head
```

```python
df2 = pd.read_csv("corona_daily.csv")
```

```python
df2.head()
```

```python
# Check Daily Dataset
print(df2.shape)
print(df2.info())
print("\n")

# Check Summary Dataset
print(df.shape)
print(df.info())
```

```python
df2.describe()
```

```python
df.describe()
```

```python
# Check missing values for both datasets
print("DAILY - Missing Values")
print(df2.isnull().sum())
print("\n")
print("SUMMARY - Missing Values")
print(df.isnull().sum())
```

```python
# Clean daily dataset
# Fill missing numeric values with 0 (no report = no cases that day)

df2["daily_new_cases"] = df2["daily_new_cases"].fillna(0)
df2["active_cases"] = df2["active_cases"].fillna(0)
df2["cumulative_total_deaths"] = df2["cumulative_total_deaths"].fillna(0)
df2["daily_new_deaths"] = df2["daily_new_deaths"].fillna(0)
```

```python
# Convert date from string to actual datetime
df2["date"] = pd.to_datetime(df2["date"])
```

```python
# Clean summary dataset
# Fill missing with 0
df.fillna(0, inplace=True)
```

```python
# Verify no more missing values
print("DAILY - Missing after cleaning")
print(df2.isnull().sum())
print("\n")
print("SUMMARY - Missing after cleaning")
print(df.isnull().sum())
```

## Data Visualization

## Chart 1 — Top 10 Countries by Total Confirmed Cases:

```python
plt.figure(figsize=(12, 6))

# Get top 10 countries
top10 = df.nlargest(10, 'total_confirmed')[['country', 'total_confirmed']]

# Create bar chart
sns.barplot(data=top10, x='total_confirmed', y='country', 
            hue='country', palette='Reds_r', legend=False)

# Labels 
plt.title('Top 10 Countries by Total COVID-19 Cases', fontsize = 16, fontweight = 'bold')

plt.xlabel('Total Confirmed Cases')
plt.ylabel('Country')

plt.tight_layout()
plt.show()
```

##  USA dominated with 84 million cases, almost double India in second place.

## Chart 2 — Global Daily Cases Over Time:

```python
plt.figure(figsize=(14, 6))

# Group by date, sum all countries
global_daily = df2.groupby('date')['daily_new_cases'].sum().reset_index()

# Line chart
plt.plot(global_daily['date'], global_daily['daily_new_cases'], 
         color='steelblue', linewidth=1)

# Labels
plt.title('Global Daily New COVID-19 Cases Over Time', fontsize=16, fontweight='bold')
plt.xlabel('Date')
plt.ylabel('Daily New Cases')

plt.tight_layout()
plt.show()
```

## 2020: Slow start, gradually rising
## 2021 mid: First big waves
## Early 2022: MASSIVE spike — that's the Omicron variant exploding globally
## 2022 mid: Dropping back down

## Chart 3: Top 10 countries by death rate (deaths/cases)

```python
plt.figure(figsize=(12, 6))

# Calculate death rate, filter countries with at least 10000 cases
df_rate = df[df['total_confirmed'] > 10000].copy()
df_rate['death_rate'] = (df_rate['total_deaths'] / df_rate['total_confirmed']) * 100

# Top 10 highest death rates
top10_deaths = df_rate.nlargest(10, 'death_rate')[['country', 'death_rate']]

sns.barplot(data=top10_deaths, x='death_rate', y='country',
            hue='country', palette='Oranges_r', legend=False)

plt.title('Top 10 Countries by COVID-19 Death Rate (%)', fontsize=16, fontweight='bold')
plt.xlabel('Death Rate (%)')
plt.ylabel('Country')

plt.tight_layout()
plt.show()
```

## Yemen had an 18% death rate — meaning 1 in 5 people who got COVID there died

## Chart 4: Total Cases by Continent

```python
plt.figure(figsize=(10, 6))

# Group by continent
continent_cases = df.groupby('continent')['total_confirmed'].sum().reset_index()
continent_cases = continent_cases.sort_values('total_confirmed', ascending=False)

sns.barplot(data=continent_cases, x='continent', y='total_confirmed',
            hue='continent', palette='Blues_r', legend=False)

plt.title('Total COVID-19 Cases by Continent', fontsize=16, fontweight='bold')
plt.xlabel('Continent')
plt.ylabel('Total Confirmed Cases')
plt.xticks(rotation=15)

plt.tight_layout()
plt.show()
```

## Europe had the MOST cases

## Chart 5 — Malaysia's Journey Over Time:

```python
plt.figure(figsize=(14, 6))

# Filter for Malaysia
malaysia = df2[df2['country'] == 'Malaysia']

plt.plot(malaysia['date'], malaysia['daily_new_cases'],
         color='green', linewidth=1.5)

# Add shading under the line
plt.fill_between(malaysia['date'], malaysia['daily_new_cases'],
                 alpha=0.3, color='green')

plt.title("Malaysia's Daily COVID-19 Cases Over Time", fontsize=16, fontweight='bold')
plt.xlabel('Date')
plt.ylabel('Daily New Cases')

plt.tight_layout()
plt.show()
```

## 2020: Very well controlled — Malaysia had strict MCO lockdowns
## Mid 2021: Delta variant hit hard — cases exploded to 25,000/day
## Late 2021: Dropped after vaccination drive
## Early 2022: Omicron wave — biggest spike at 33,000/day
## Mid 2022: Finally declining


## Chart 6: Correlation between total tests and total cases

```python
plt.figure(figsize=(10, 7))

# Filter out countries with 0 tests
df_test = df[df['total_tests'] > 0]

sns.scatterplot(data=df_test, x='total_tests', y='total_confirmed',
                hue='continent', s=100)

plt.title('Correlation: Total Tests vs Total Cases by Country', 
          fontsize=16, fontweight='bold')
plt.xlabel('Total Tests')
plt.ylabel('Total Confirmed Cases')

plt.tight_layout()
plt.show()
```

## China (Asia) — did 850 million tests but reported very few cases — suspicious underreporting!
